##### 1. Why did you pick the specific chart?



> I chose the count plot because the target variable (Machine failure) is categorical and binary in nature (0 = No Failure, 1 = Failure). A count plot is the most suitable visualization to understand the distribution of classes in classification problems. It helps in identifying whether the dataset is balanced or imbalanced, which is an important factor for model selection and evaluation. Understanding class distribution also helps in deciding whether techniques like SMOTE or class weighting are required during model training.


##### 2. What is/are the insight(s) found from the chart?

> From the chart, it was observed that the dataset is imbalanced, with a significantly higher number of non-failure cases compared to failure cases. This indicates that most machines are operating normally, while only a small portion of records represent machine failure. This imbalance is an important insight because it can affect model performance, making the model biased toward predicting the majority class if not handled properly. Therefore, techniques like class balancing (SMOTE or class weights) may be required to improve model accuracy for failure prediction.

Are there any insights that lead to negative growth? Justify with specific reason.

> Yes, the insights gained from this chart can create a strong positive business impact. > However, there is also a potential negative aspect due to class imbalance in the dataset. Since failure cases are much fewer compared to non-failure cases, a machine learning model trained on this data may become biased toward predicting "no failure" more often. This can lead to missed failure predictions (false negatives), which is risky in an industrial environment like Tata Steel, as undetected failures can cause serious equipment damage, safety issues, and production losses. Therefore, proper handling of imbalance is necessary to avoid negative business impact.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0):
There is no significant difference in process temperature between machines that fail and those that do not fail.

Alternate Hypothesis (H1):
There is a significant difference in process temperature between machines that fail and those that do not fail.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

# Split data based on target variable
group1 = df[df['Machine failure'] == 0]['Process temperature [K]']
group2 = df[df['Machine failure'] == 1]['Process temperature [K]']

# Perform Independent T-test
stat, p_value = ttest_ind(group1, group2)

print("T-Statistic:", stat)
print("P-Value:", p_value)
````

I have used the Independent Sample T-test (Two-sample t-test) to obtain the p-value.

##### Why did you choose the specific statistical test?

I chose the Independent T-test because we are comparing the mean of a continuous numerical variable (Process Temperature) between two independent groups: machines that failed and machines that did not fail. This test helps determine whether the difference in means between the two groups is statistically significant or occurred by chance. Since the target variable is binary and the feature is continuous, the Independent T-test is appropriate for this hypothesis testing scenario.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Base model
log_reg = LogisticRegression(max_iter=1000)

# Hyperparameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l2'],
    'solver': ['lbfgs']
}

# GridSearchCV with 5-fold cross validation
grid_search = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1
)

# Fit model
grid_search.fit(X_train, y_train)

# Best model
best_model_2 = grid_search.best_estimator_

print("Best Parameters:", grid_search.best_params_)

# Predict on the model

y_pred_2 = best_model_2.predict(X_test)

print("Model-2 Predictions Completed")

from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred_2))
print("\nClassification Report:\n", classification_report(y_test, y_pred_2))
````

It also helps in selecting the best hyperparameter combination that gives the highest performance based on the chosen evaluation metric, such as F1-score. This technique reduces the risk of overfitting and improves the model’s ability to generalize well on unseen data, making it suitable for classification problems like machine failure prediction.

Precision measures how many of the predicted failures are actually true failures. From a business perspective, high precision means fewer false alarms, which reduces unnecessary maintenance costs and avoids disrupting production for machines that are actually working fine.

Recall is the most critical metric in this project because it measures how many actual machine failures are correctly identified by the model. High recall means the model successfully detects most failure cases, which is very important in industry since missing a failure can lead to unexpected breakdowns, production loss, and high repair costs.

F1-score is the balance between precision and recall. It provides a single metric that considers both false positives and false negatives. In a business context, it ensures that the model is not only accurate in detecting failures but also not generating too many false alarms.

Overall, the business impact of the ML model is significant because it enables predictive maintenance. By identifying potential machine failures in advance, the company can reduce downtime, optimize maintenance scheduling, lower operational costs, and improve production efficiency. The model helps shift maintenance strategy from reactive to proactive, which is highly valuable in manufacturing industries.
